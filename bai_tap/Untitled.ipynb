import thứ viện

```python

import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Du_lieu.csv")
data
```

```python
tmp_1 =[]
tmp_2 =[]
for d in data['Gia_cu']:
    d = d.replace("đ" , "")
    d = d.replace("." ,"")
    tmp_1.append(int(d))
for d in data['Gia_moi']:
    d = d.replace("đ" , "")
    tmp_2.append(int(d.replace("." , "")))
       
data['Gia_cu'] = tmp_1
data["Gia_moi"] = tmp_2
data
```

```python
discount = 1 - data['Gia_moi'] / data["Gia_cu"]
data["Giam_gia"] = discount
data
```

các sản phẩm có giảm giá trên 20%
